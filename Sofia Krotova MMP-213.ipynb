# МЕРЫ РАЗБРОСА

```python
import pandas as pd
cosmetics = pd.read_pickle('cosmeticsVars.pkl')
cosmetics.head()
```

```python
#Посмотрим визуализацию колонки

cosmetics_cost.plot(kind='bar')
```

```python
#Найдем размах

cosmetics_cost['Rank'].max()
```

```python
#Найдем размах нашей переменной

cosmetics_cost['Rank'].max() - cosmetics_cost['Rank'].min()
```

```python
#Найдем интерквартильный размах

cosmetics_cost.describe()['Rank']['25%']
```

```python
#Найдем интерквартильный размах другим способом

np.percentile(cosmetics_cost['Rank'], 25)
```

```python
#Посчитаем интерквартильный размах

np.percentile(cosmetics_cost['Rank'], 75) - np.percentile(cosmetics_cost['Rank'], 25)
```

```python
#Посчитаем и найдем дисперсию, отклонения

cosmetics_mean = cosmetics_cost['Rank'].mean()
deviations = cosmetics_cost['Rank'] - cosmetics_mean
print (deviations)
```

```python
#Найдем отклонения, возведенные в квадрат, отрицательные числа не пропали

cosmetics_mean = cosmetics_cost['Rank'].mean()
deviations = cosmetics_cost['Rank'] - cosmetics_mean
print (deviations)
sq_deviations = deviations ** 2
print(sq_deviations)
```

```python
#Найдем дисперсию и среднееквадратичное отклонение

cosmetics_mean = cosmetics_cost['Rank'].mean()
deviations = cosmetics_cost['Rank'] - cosmetics_mean
print (deviations)
sq_deviations = deviations ** 2
print(sq_deviations)
variance_samp = sq_deviations.sum() / (len(cosmetics_cost) - 1)
std_samp = np.sqrt(variance_samp)
print(variance_samp, std_samp)
```

```python
#Найдем дисперсию среднеквадратичного отклонения генеральной совокупности

variance_pop = sq_deviations.sum() / (len(cosmetics_cost))
std_pop = np.sqrt(variance_pop)
print(variance_pop, std_pop)
```

```python
#Найдем дисперсию и среднеквадратичное отклонение для выборки с помощью специальной функции

print(cosmetics_cost['Rank'].var())
print(cosmetics_cost['Rank'].std())
```

```python
#Найдем дисперсию и среднеквадратичное отклонение для генеральной совокупности с помощью специальной функции

print(cosmetics_cost['Rank'].var(ddof=0))
print(cosmetics_cost['Rank'].std(ddof=0))
```

```python
#Сгруппируем по среднеквадратичному отклонению для каждого уровня

cosmetics.groupby('Price Category')['Rank'].std()
```

```python
#Сгруппируем по среднеквадратичному отклонению для каждого уровня через agg

cosmetics.groupby('Price Category')['Rank'].agg(['mean', 'std'])
```

```python
#Пивотальная таблица

def range_values(x):
    return x.max() - x.min()
cosmetics.pivot_table(index='Rank',
                     values=['Price'],
                     aggfunc=['var', 'std', range_values])
```

# ВЫБРОСЫ

```python
import pandas as pd
import numpy as np

cosmetics = pd.read_pickle('cosmeticsVars.pkl')
cosmetics.head()
```

```python
#Посмотрим на основные описательные статистики, минимальное кол-во 3, макс 370

cosmetics['Price'].describe()
```

```python
#Найдем интерквантильный размах

q3 = np.percentile(cosmetics['Price'], 75)
q1 = np.percentile(cosmetics['Price'], 25)
print(q3, q1)
```

```python
#Найдем интерквантильный размах

iqr = q3 - q1
print(iqr)
```

```python
#Найдем границу сверху и границу снизу, слева от медианы выбросов не будет, справа будут

iqr_threshold_bottom = q1 - 1.5 * iqr
iqr_threshold_top = q3 + 1.5 * iqr
print(iqr_threshold_bottom, iqr_threshold_top)
```

```python
#Отфильтруем датасет

cosmetics[cosmetics['Price'] > iqr_threshold_top]
```

```python
#Отфильтруем датасет, найдем, сколько наблюдений будем считать выбросами

cosmetics[cosmetics['Price'] > iqr_threshold_top].shape
```

```python
#Найдем по какому рейтингу сгруппированы эти наблюдения

cosmetics[cosmetics['Price'] > iqr_threshold_top]['Rank'].value_counts()
```

```python
#Найдем по какой категории сгруппированы эти наблюдения

cosmetics[cosmetics['Price'] > iqr_threshold_top]['Label'].value_counts()
```

```python
#Найдем выбросы через среднеквадратичное отклонение

mean = cosmetics['Price'].mean()
std = cosmetics['Price'].std()
std_threshold_bottom = mean - 2.5 * std
std_threshold_top = mean + 2.5 * std
print(std_threshold_bottom, std_threshold_top)
```

```python
#Берем верхний порог, так как на нижнем нет нужных значений

cosmetics[cosmetics['Price'] > std_threshold_top].shape
```

```python
#Отфильтруем значения и посмотрим, как будут выглядеть меры центральной тенденции в выборках, где нет и есть выбросы

iqr_no_outliers = cosmetics[cosmetics['Price'] <= iqr_threshold_top]
std_no_outliers = cosmetics[cosmetics['Price'] <= std_threshold_top]
```

```python
#Все три значения отличаются, первое оригинальное

print(cosmetics['Price'].mean())
print(iqr_no_outliers['Price'].mean())
print(std_no_outliers['Price'].mean())
```

```python
#Посмотрим, как меняется медиана, она оказалась более устойчивой

print(cosmetics['Price'].median())
print(iqr_no_outliers['Price'].median())
print(std_no_outliers['Price'].median())
```

# КОРРЕЛЯЦИЯ

```python
import pandas as pd
import numpy as np

cosmetics = pd.read_pickle('cosmeticsVars.pkl')
cosmetics.head()
```

```python
#Мало значений для сравнения корреляции, но посмотрим, какие товары имеют высокий рейтинг и цену

rank_mean = cosmetics.groupby(cosmetics['Rank'])['Price'].mean()
price_sum = cosmetics.groupby(cosmetics['Rank'])['High rating'].mean()
```

```python
rank_price = pd.concat([rank_mean, price_sum], axis=1)
```

```python
first_five = rank_price.iloc[:5]
first_five
```

```python
first_five['X - Mx'] = first_five['Price'] - first_five['Price'].mean()
first_five['Y - My'] = first_five['High rating'] - first_five['High rating'].mean()
```

```python
first_five
```

```python
#Плохо вышла корреляция из-за недостатка данных

SSx = (first_five['X - Mx'] ** 2).sum
SSy = (first_five['Y - My'] ** 2).sum
SP = (first_five['X - Mx'] * first_five['Y - My']).sum()
print (SSx, SSy, SP)
```

```python
#По графику корреляции не наблюдается

rank_price.plot('Price', 'High rating', kind='scatter')
```
